```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv("data/epa-sea-level.csv")

    # Create scatter plot
    year = df["Year"]
    sea_lvl = df["CSIRO Adjusted Sea Level"]
    plt.scatter(x=year, y=sea_lvl, label='Observed Sea Level')

    # Create first line of best fit
    yrs_predict = np.array(list(range(1880, 2051)))
    slope, intercept, r_value, p_value, std_err = linregress(year, sea_lvl)
    plt.plot(yrs_predict, intercept + slope*yrs_predict, "r", label="Fitted Line (All Data)")

    # Create second line of best fit
    BM = df.Year >= 2000
    df_2k = df[BM].copy()
    slope, intercept, r_value, p_value, std_err = linregress(df_2k["Year"], df_2k["CSIRO Adjusted Sea Level"])
    yrs_predict = np.array(list(range(2000, 2051)))
    plt.plot(yrs_predict, intercept + slope*yrs_predict, "orange", label="Fitted Line (2000-2013)")

    # Add labels and title
    plt.title("Rise in Sea Level")
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")
    plt.legend()
    
    # Save plot and return data for testing (DO NOT MODIFY)
    # plt.savefig('sea_level_plot.png')
    # return plt.gca()
```

```python
draw_plot()
```
